```python
import shutil
import os
for r in df_f.iterrows():
    i = r[1]['VoxCeleb1 ID']
    print(i)
    f = r[1]['VGGFace1 ID']
    print(f)
    break
```

```python
import shutil
import os
for r in df_f.iterrows():
    i = r[1]['VoxCeleb1 ID']
    f = r[1]['VGGFace1 ID']
    
    if os.path.exists("voxCeleb-Dataset\\fbank\\fbank - Copy\\" + i) == True and os.path.exists("voxCeleb-Dataset\\VGG_ALL_FRONTAL\\VGG_ALL_FRONTAL - Copy\\"+f) == True:
        s_o = "voxCeleb-Dataset\\fbank\\fbank - Copy\\" + i
        s_t = "voxCeleb-Dataset\\females\\voices\\" + i
        #shutil.copyfile(s_o,s_t)
        os.rename(s_o,s_t)
        
        f_o = "voxCeleb-Dataset\\VGG_ALL_FRONTAL\\VGG_ALL_FRONTAL - Copy\\"+f
        f_t = "voxCeleb-Dataset\\females\\faces\\"+f
        #shutil.copyfile(f_o,f_t)
        os.rename(f_o,f_t)
        print(r , "Done")
```

```python
import glob
l = glob.glob("voxCeleb-Dataset\\females\\faces\\*")
l2 = glob.glob("voxCeleb-Dataset\\females\\voices\\*")
print(len(l) , len(l2))
```
